```python
import pandas as pd
import numpy as np

data = pd.read_csv('1. Weather Data.csv')

# read the first 5 rows of the data
data.head()
# # last 5 rows of the data
# data.tail()
# # number of rows and columns
# data.shape
```

```python
# Find all records where the weather was exactly clear.
data[data['Weather'] == 'Clear']
```

```python
# Find the number of times the wind speed was exactly 4 km/hr.
data[data['Wind Speed_km/h'] == 4].shape[0]
```

```python
# Check if there are any NULL values present in the dataset.
data.isnull().sum()
```

```python
# Rename the column "Weather" to "Weather_Condition."
data.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
list(data.columns)
```

```python
# What is the mean visibility of the dataset?
data['Visibility_km'].mean()
```

```python
# Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.
data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)].shape[0]
```

```python
# Mean value of each column for each weather condition
weather_conditions = data['Weather_Condition'].str.split(',', expand=True).stack().reset_index(level=1, drop=True).rename('Weather_Condition')
data = data.drop('Weather_Condition', axis=1).join(weather_conditions)

numeric_columns = data.select_dtypes(include=[np.number]).columns

# Group by 'Weather_Condition' and calculate the mean for numeric columns
data.groupby('Weather_Condition')[numeric_columns].mean()
```

```python
# Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
data[(data['Weather_Condition'] == 'Clear') & (data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40)]
```

```python
# Find the number of weather conditions that include snow.
data['Weather_Condition'].str.contains('Snow').sum()
```
